```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets

df = pd.DataFrame(datasets.load_iris().data)
df.columns = ['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width']
df.head(2)
```

```python
x = np.linspace(1, 100, 100)
y = x**2
plt.plot(x, y)
```

```python
# print multiple plots
x = np.linspace(1, 10, 100)
y1 = x**2
y2 = x**3
y3 = x**4

plt.subplot(1, 3, 1)
plt.plot(x, y1)
plt.subplot(1, 3, 2)
plt.plot(x, y2)
plt.subplot(1, 3, 3)
plt.plot(x, y3)
```

```python
fig, axes = plt.subplots(figsize=(8, 4), nrows=1, ncols=3, sharex=True)
plt.tight_layout()  # avoid plot overlap
axes[0].set_title('Plot 1')
axes[1].set_title('Plot 2')
axes[2].set_title('Plot 3')

# plot using axes.plot
x1 = np.linspace(1, 10, 50)
y1 = np.sin(x1)
y2 = np.cos(x1)
y3 = np.tan(x1)
axes[0].plot(x1, y1, ls='-.', marker='o',
             markerfacecolor='y', markeredgecolor='y')
axes[1].plot(x1, y2)
axes[2].plot(x1, y3)
```
