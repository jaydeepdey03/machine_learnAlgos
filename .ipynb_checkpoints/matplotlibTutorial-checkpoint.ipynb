```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(1, 100, 100)
y = x**2
plt.plot(x, y)
```

```python
# print multiple plots
x = np.linspace(1, 100, 10000)
y1 = x**2
y2 = x**3
y3 = x**4

plt.subplot(1, 3, 1)
plt.plot(x, y1)
plt.subplot(1, 3, 2)
plt.plot(x, y2)
plt.subplot(1, 3, 3)
plt.plot(x, y3)
```
