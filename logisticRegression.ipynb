```python
from sklearn import datasets
iris = datasets.load_iris()
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
```

```python
print(iris.DESCR)
```

```python
print(pd.DataFrame(iris.data))
```

```python
Y = (iris.target == 2).astype(int)
X = iris.data[:, 3: ]
```

```python
model = LogisticRegression()
model.fit(X, Y)
print(model.predict(([[2.6]])))
```

```python
X_new = np.linspace(0, 3, 1000).reshape(-1,1)
X_new1 = np.linspace(0, 3, 1000)
X_prob = model.predict_proba(X_new)
X_new
```

```python
plt.plot(X_new, X_prob[:, 1], label="virginica", color="blue")
plt.show()
```
